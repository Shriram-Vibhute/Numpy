```python
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)

xx, yy = np.meshgrid(x, y)
```

```python
zz = xx ** 2 + yy ** 2
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()

plt.style.use('fivethirtyeight')

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx, yy, zz, cmap = 'viridis')
plt.colorbar(p)
```
